```python
import matplotlib.pyplot as plt
import numpy as np


from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
%matplotlib inline


```

```python
#storing every player's stat's in a dictionary called teamDict
#each item in teamDict is a dictionary containing a player's stats. 
#teamDict["playerName"] = {stat1: value, stat2: value....}
teamDict = dict()
with open("stats.txt", 'r', encoding="ISO-8859-1") as f:
	for line in f:
		pairs = line.split(',')
		playerDict = dict()
		for i in pairs:
			i = i.replace("'", "").replace("{", "").replace("}", "").strip()
			li = i.split(":")
			for j in range(len(li)):
				li[j] = li[j].replace(' ', '')
			print(i)
			playerDict[li[0]] = int(li[1]) if li[1].isnumeric() else li[1]
		teamDict[playerDict['name']] = playerDict
```

```python
#top points for each position according to FPL
topForwardPoints = 242
topMidfielderPoints = 244
topDefenderPoints = 171
topGoalkeeperPoints = 186
```

```python
#important stats for Forwards
#positive
forwardGoals = 23
forwardAssists = 14
forwardPasses = 1301

#negative
forwardChancesMissed = 21
forwardYellowCards = 7
forwardRedCards = 1
forwardPenaltiesMissed = 1

```

```python
#important stats for Midfielders
#positive
midfielderGoals = 18
midfielderAssists = 12
midfielderCleanSheets = 14
midfielderPasses = 2785

#negative
midfielderYellowCards = 12
midfielderRedCards = 1
midfielderPenaltiesMissed = 2
```

```python
#important stats for Defenders
#positive
defenderGoals = 8 
defenderAssists = 11
defenderCleanSheets = 17

#negative
defenderYellowCards = 11
defenderRedCards = 2
defenderGoalsConceded = 71
defenderOwnGoals = 2
```

```python
#important stats for Goalkeepers
#positive
goalkeeperCleanSheets = 19 
goalkeeperPenaltiesSaved = 1
goalkeeperSaves = 166

#negative
goalkeeperYellowCards = 3
goalkeeperRedCards = 1
goalkeeperGoalsConceded = 74

```

```python
def getSentiment(playerName):

	sid_obj = SentimentIntensityAnalyzer()
	
	neutral_count = 0
	positive_count = 0
	negative_count = 0

	with open(playerName + ".txt", 'r') as f:
		for tweet in f:
			sentiment_dict = sid_obj.polarity_scores(tweet)
	
	
			# decide sentiment as positive, negative and neutral
			if sentiment_dict['compound'] >= 0.05 :
				positive_count += 1
		
			elif sentiment_dict['compound'] <= - 0.05 :
				negative_count += 1
		
			else:
				neutral_count += 1
	
		print("neutral: " + str(neutral_count))
		print("positive: " + str(positive_count))
		print("negative: " + str(negative_count))
	
		sentDict = dict()
		sentDict['neutral'] = neutral_count
		sentDict['positive'] = positive_count
		sentDict['negative'] = negative_count
	
		return sentDict
```

```python
#appending another dictionary to each player
#this dictionary will contain the count of negative, positive, and neutral tweets.
for player in teamDict.keys():
	print(player.replace('-', ' '))
	teamDict[player]["sentiment"] = getSentiment(player.replace('-', ''))
	print(teamDict[player]["sentiment"])

```

```python
def showSentPlot(playerName):

    fig = plt.figure()
    neg = teamDict[playerName]["sentiment"]["negative"]
    neut = teamDict[playerName]["sentiment"]["neutral"]
    pos = teamDict[playerName]["sentiment"]["positive"]
    ax = fig.add_axes([0,0,1,1])
    sent = ['negative', "neutral", "positive"]
    counts = [neg, neut, pos]

    total = neg+neut+pos
    print("negatve percent: ", round(neg/total, 2) )
    print("neutral percent: ", round(neut/total, 2) )
    print("positive percent: ", round(pos/total, 2) )

    ax.bar(sent, counts)
    plt.show()
```

```python
def compareStats(playerName):
    position = teamDict[playerName]["position"]
    if position == "Goalkeeper":
        
        totalPoints = teamDict[playerName]['total_points']
        cleanSheets = teamDict[playerName]['clean_sheets']
        penaltiesSaved = teamDict[playerName]['PK_saves']
        saves = teamDict[playerName]['saves']
        yellowCards = teamDict[playerName]['yellow_cards']
        redCards = teamDict[playerName]['red_cards']
        goalsConceded = teamDict[playerName]['goals_conceded']
        
        width = .3
        #-----------------------------------------------------------------
        #plot one for positive stats
        categories = ["Total Points", "Clean sheets", "Penalties saved", "Saves"]
        
        posData = [[totalPoints, cleanSheets, penaltiesSaved, saves],
        [topGoalkeeperPoints, goalkeeperCleanSheets, goalkeeperPenaltiesSaved, goalkeeperSaves]]
        X1 = np.arange(4)
        X2 = [i+width for i in X1]
        
        plt.bar(X1, posData[0], width, label=playerName)
        plt.bar(X2, posData[1], width, label="Top in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)
        plt.show()
        
        #-------------------------------------------------------------------
        #plot 2 for negative stats
        categories = ["Goals conceded", "Yellow Cards", "Red Cards"]
        
        negData = [[goalsConceded, yellowCards, redCards],
        [goalkeeperGoalsConceded, goalkeeperYellowCards, goalkeeperRedCards]]
        X1 = np.arange(3)
        X2 = [i+width for i in X1]
        
        plt.bar(X1, negData[0], width, label=playerName)
        plt.bar(X2, negData[1], width, label="Worst in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)   
        plt.show()
    
    elif position == "Defender":
        
        totalPoints = teamDict[playerName]['total_points']
        cleanSheets = teamDict[playerName]['clean_sheets']
        goals = teamDict[playerName]['goals']
        assists = teamDict[playerName]['assists']
        yellowCards = teamDict[playerName]['yellow_cards']
        redCards = teamDict[playerName]['red_cards']
        goalsConceded = teamDict[playerName]['goals_conceded']
        ownGoals = teamDict[playerName]["own_goals"]
        
        width = .3
        #-----------------------------------------------------------------
        #plot one for positive stats
        categories = ["Total Points", "Goals", "Assists", "Clean sheets"]
        
        posData = [[totalPoints, goals, assists, cleanSheets],
        [topDefenderPoints, defenderGoals, defenderAssists, defenderCleanSheets]]
        X1 = np.arange(4)
        X2 = [i+width for i in X1]
        
        plt.bar(X1, posData[0], width, label=playerName)
        plt.bar(X2, posData[1], width, label="Top in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)
        plt.show()
        
        #-------------------------------------------------------------------
        #plot 2 for negative stats
        categories = ["Goals conceded","Own goals", "Yellow Cards", "Red Cards"]
        
        negData = [[goalsConceded,ownGoals, yellowCards, redCards],
        [defenderGoalsConceded, defenderOwnGoals, defenderYellowCards, defenderRedCards]]
        X1 = np.arange(4)
        X2 = [i+width for i in X1]
        
        plt.bar(X1, negData[0], width, label=playerName)
        plt.bar(X2, negData[1], width, label="Worst in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)   
        plt.show()
    
    elif position == "Midfielder":

        totalPoints = teamDict[playerName]['total_points']
        goals = teamDict[playerName]['goals']
        assists = teamDict[playerName]['assists']
        yellowCards = teamDict[playerName]['yellow_cards']
        redCards = teamDict[playerName]['red_cards']
        passes = teamDict[playerName]['passes']

        
        width = .3
        #-----------------------------------------------------------------
        #plot one for positive stats
        categories = ["Total Points","Goals", "Assists"]
        
        posData = [[totalPoints, goals, assists],
        [topMidfielderPoints, midfielderGoals, midfielderAssists]]
        X1 = np.arange(3)
        X2 = [i+width for i in X1]

        
        plt.bar(X1, posData[0], width, label=playerName)
        plt.bar(X2, posData[1], width, label="Top in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)
        plt.show()

        categories = ["Passes"]
        
        passData = [[passes],[midfielderPasses]]
        X1 = np.arange(1) 
        X2 = [i+width for i in X1] 
        
        plt.bar(X1, posData[0], width, label=playerName)
        plt.bar(X2, posData[1], width, label="Top in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)
        plt.show()
        
        #-------------------------------------------------------------------
        #plot 2 for negative stats
        categories = ["Yellow Cards", "Red Cards"]
        
        negData = [[yellowCards, redCards],
        [midfielderYellowCards, midfielderRedCards]]
        X1 = np.arange(2)
        X2 = [i+width for i in X1]
        
        plt.bar(X1, negData[0], width, label=playerName)
        plt.bar(X2, negData[1], width, label="Worst in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)   
        plt.show()
    
    elif position == "Forward":
        
        totalPoints = teamDict[playerName]['total_points']
        goals = teamDict[playerName]['goals']
        assists = teamDict[playerName]['assists']
        yellowCards = teamDict[playerName]['yellow_cards']
        redCards = teamDict[playerName]['red_cards']
        chancesMissed = teamDict[playerName]['big_chances_missed']
        passes = teamDict[playerName]['passes']

        
        width = .3
        #-----------------------------------------------------------------
        #plot one for positive stats
        categories = ["Total Points", "Goals", "Assists"]
        
        posData = [[totalPoints,goals, assists],
        [topForwardPoints, forwardGoals, forwardAssists]]
        X1 = np.arange(3)
        X2 = [i+width for i in X1]
        
        plt.bar(X1, posData[0], width, label=playerName)
        plt.bar(X2, posData[1], width, label="Top in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)
        plt.show()
        
        categories = ["Passes"]
        
        passData = [[passes],[forwardPasses]]
        X1 = np.arange(1) 
        X2 = [i+width for i in X1] 
        
        plt.bar(X1, posData[0], width, label=playerName)
        plt.bar(X2, posData[1], width, label="Top in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)
        plt.show()
        #-------------------------------------------------------------------
        #plot 2 for negative stats
        categories = ["Big chances missed", "Yellow Cards", "Red Cards"]
        
        negData = [[chancesMissed, yellowCards, redCards],
        [forwardChancesMissed,forwardYellowCards, forwardRedCards]]
        X1 = np.arange(3)
        X2 = [i+width for i in X1]
        
        plt.bar(X1, negData[0], width, label=playerName)
        plt.bar(X2, negData[1], width, label="Worst in position")
        
        plt.xlabel("Categories")
        plt.ylabel("Number")
        plt.title("Player performance")
        plt.legend()
        
        plt.xticks(X1+width/2, categories)   
        plt.show()
```
